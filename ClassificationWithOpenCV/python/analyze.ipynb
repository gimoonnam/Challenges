```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory = '../data/'
names = ['moon','circle','blob']

def ImportData(name, size,noise):
    Inputfile = os.path.join(directory,'%s_%s_%s.txt'% (name,size,noise))
    print(Inputfile)

    with open(Inputfile) as f:
        lines = f.readlines()
    data = []
    for l in lines:
        values = l.split()
        data.append([float(values[0]),float(values[1]),float(values[2])]);
    
    data = np.array(data)
    return data


def PlotData(data):
    idx0 = np.where(data[:,2]==0)
    idx1 = np.where(data[:,2]==1)
    plt.scatter(data[idx0,0],data[idx0,1])
    plt.scatter(data[idx1,0],data[idx1,1])
        
```

```python
mLl = ImportData('moon','large','low')
mLh = ImportData('moon','large','high')
```

```python
PlotData(mLl)
```

```python
PlotData(mLh)
```
